```python
# 1. Importation des librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Pour affichage correct
%matplotlib inline
sns.set(style="whitegrid")

# 2. Chargement du dataset
df = pd.read_csv("./house_prices.csv")

# 3. Aperçu général du dataset
print("Dimensions du dataset :", df.shape)
df.head()
```

```python
# 4. Informations générales
df.info()
df.describe(include="all")
```

```python
# 4. Nettoyage et préparation des données des données
# Suppression des colonnes inutiles pour l’analyse de base
df.drop(columns=["id", "date"], inplace=True)

# Conversion de certaines colonnes
df["waterfront"] = df["waterfront"].replace({'Y': 1, 'N': 0}) if df["waterfront"].dtype == 'object' else df["waterfront"]
df["condition"] = df["condition"].astype(str)  # si elle est catégorielle

# Vérification des valeurs manquantes
missing = df.isnull().sum()
missing[missing > 0]
```

```python
# 6. Analyse de la variable cible (price)
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title("Distribution des prix")
plt.xlabel("Prix")
```

```python
# 7. Corrélation entre variables numériques
plt.figure(figsize=(14, 12))
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matrice de corrélation")
```

```python
# 8. Visualisations des boites à moustaches

# Prix par nombre de chambres
plt.figure(figsize=(8, 6))
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title("Prix par nombre de chambres")

# Prix par grade
plt.figure(figsize=(8, 6))
sns.boxplot(x='grade', y='price', data=df)
plt.title("Prix par grade")
```

```python
# 9. Carte : localisation des maisons
plt.figure(figsize=(10, 6))
sns.scatterplot(x="long", y="lat", hue="price", size="sqft_living", data=df, palette="viridis", alpha=0.6)
plt.title("Carte des maisons (couleur = prix, taille = surface)")
```

```python
# 10. Focus sur les maisons rénovées
plt.figure(figsize=(8, 6))
sns.boxplot(x=(df["yr_renovated"] > 0), y="price", data=df)
plt.xticks([0, 1], ['Non rénovées', 'Rénovées'])
plt.title("Prix des maisons selon rénovation")
```

```python
# 11. Pour terminer : voici le top 10 des maisons les plus chères
top10 = df.sort_values(by="price", ascending=False).head(10)

# Création d'une nouvelle colonne pour les étiquettes
top10["label"] = top10["zipcode"].astype(str) + " - " + top10["sqft_living"].astype(str) + " sqft"

# Création de la figure
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x="price", y="label", data=top10, palette="magma")

# Ajout des étiquettes de prix sur les barres
for i, (price, label) in enumerate(zip(top10["price"], top10["label"])):
    plt.text(price + 10000, i, f"${int(price):,}", va='center')

# Titres et labels
plt.title("Top 10 des maisons les plus chères")
plt.xlabel("Prix ($)")
plt.ylabel("Emplacement - Surface")
plt.tight_layout()
plt.show()
```
